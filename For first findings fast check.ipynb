```python
import pandas as pd
import json
```

```python
gene_set = pd.read_excel('Different genes to check.xlsx')#Считываю список генов на проверку
```

```python
with open('Expressions.json', "r") as f: #читаю json документ экспрессий пациента
    expr = json.load(f)
    
final = expr['expressions']
```

```python
df_to_report = pd.DataFrame(columns = ['Gene_name', 'Expression_value', 'Treatment_or_prognosis', 'PMID']) 
#Создаю дата фрейм, в который будет записываться финальный результат
```

```python
for i in gene_set.index: #пробегаюсь по генам, указанным в списке генов для проверки
    for l in final: #пробегаюсь по генам из списка генов пациента
        if l['geneId'] == gene_set.loc[i, 'Gene_name'] and str(l['value']['cohortRelativeValue']['level']).lower() == gene_set.loc[i, 'Expression_value']: 
        #условие - совпадение названия гена и уровня его экспрессии по когорте
            df_to_report = df_to_report.append(gene_set.loc[i]) 
            #если условие выполнилось - добавляю строку из информации о генах в мой новый датафрейм
```

```python
sheet_name = str(input('Введите номер репорта ')) #Прошу дать название новому листу в соответствии с названием репорта
df_to_report.to_excel('Check the report.xlsx', index = False, sheet_name = sheet_name) #сохраняю в мой документ
```
